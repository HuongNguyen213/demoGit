```python
import pandas as pd 


url = "https://raw.githubusercontent.com/anhdung98/diem_thi_2022/main/diem_thi_thpt_2022.csv"
df = pd.read_csv(url) # Lấy dữ liệu từ file work_data.csv đê phân tích và đặt tên dữ liệu đó là dataset
df
```

### Create a data frame df_HCM containing the list of examinees at Ho Chi Minh City Examination Council, whose first 2 digits of the registration number (sbd) are "02".

```python
df_HCM = df[df['sbd'].astype(str).str.startswith('2')]
print(df_HCM)
```

### Print to the screen the number of examinees who have taken part in the exam at Ho Chi Minh City Examination Council.

```python
df_hcme = df_HCM.count()
print(df_hcme)
```

### Print to the screen the number of examinees who have the sum score of mathematics (toan) and physics (vat_li) lower than 5 at the Ho Chi Minh City Examination Council. Display on the screen the list of these examinees.

```python
df_hcmMark = df_HCM[(df_HCM['toan'] + df_HCM['vat_li'])<5]
print(df_hcmMark)

print("---------------------------------------------------------------------------------")
df_count = df_hcmMark.count()
print("the number of examinees who have the sum score of mathematics (toan) and physics (vat_li) lower than 5 at the Ho Chi Minh City Examination Council.", df_count)
```

### Calculate, then compare, the average (mean) foreign languages scores of
Ho Chi Minh City Examination Council with those of Quang Ninh (if it is higher/ lower or equal). Display the result on the screen.

```python
hcm = df_HCM[['ngoai_ngu']].mean()
hcm.iloc[0]
```

```python
df_QN = df[df['sbd'].astype(str).str.startswith('17')]
qn = df_QN[['ngoai_ngu']].mean()
qn.iloc[0]
```

```python
if hcm.iloc[0] > qn.iloc[0]:
    print("the average foreign languages scores of HCM is higher than QN")
elif hcm.iloc[0] < qn.iloc[0]:
    print("the average foreign languages scores of HCM is lower than QN")
else:
    print("the average foreign languages scores of HCM is equal than QN")
```
