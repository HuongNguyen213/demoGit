```python
from collections import Counter
def Frequent_word():
    # Read the file
    f =  open("data/text1.txt", encoding="utf8")
    data = f.read()

    # Count the occurrences of each word
    words = data.split()
    word_counts = Counter(words)

    # Print the nth most frequent words and their counts
#     frequency_word = int(input('\nEnter the n most frequency: '))
    print(' words with the most frequency are:\n')

    for i,(word, count) in enumerate( word_counts.most_common()):

        print(f" {i+1}.{word} : {count}")
    print('\n')
Frequent_word()
```
