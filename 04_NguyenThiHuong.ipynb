```python
# a. Declare libraries
import pandas as pd
import matplotlib.pyplot as plt

# b. Read data from file
url = "https://raw.githubusercontent.com/anhdung98/diem_thi_2022/main/diem_thi_thpt_2022.csv"
df = pd.read_csv(url)

print("c. Display on the screen the last 8 rows of data from file")
print("__________________________________________________________________________________________________________________")

print(df.tail(8))
print("__________________________________________________________________________________________________________________")


print("d. Print to the screen the number of examniees who have taken part in the physics(vat_li) exam ")
print("__________________________________________________________________________________________________________________")

print(df.vat_li.count())
print("__________________________________________________________________________________________________________________")


# e. Create a data frame df_TN containing the list of examinees at Thai Nguyen Examination Council - who have the first 2 digits of the registration number(sbd) are "12"
df_TN = df[df['sbd'].astype(str).str.startswith('17')]
print("f. print a data frame df_TN containing the list of examinees at Thai Nguyen Examination Council - who have the first 2 digits of the registration number(sbd) are "'12'"")
print("__________________________________________________________________________________________________________________")
print(df_TN)
print("__________________________________________________________________________________________________________________")


print("g.  Calculate, then compare, the average (mean) literature scores of Thai Nguyen with those of nationwide (if it is higher/lower or equal). Display the result on the screen ")
print("__________________________________________________________________________________________________________________")
mean_TN = df_TN[['ngu_van']].mean()
mean_NW = df[['ngu_van']].mean()

if mean_TN.iloc[0] > mean_NW.iloc[0]:
    print("the average  literature scores of Thai Nguyen is higher than nationwide")
elif mean_TN.iloc[0] < mean_NW.iloc[0]:
    print("the average  literature scores of Thai Nguyen is lower than nationwide")
else:
    print("the average  literature scores of Thai Nguyen is equal than nationwide")

print("__________________________________________________________________________________________________________________")


print("h. Display on the screen the list of examinees who have both scores of mathematics (toan) and literature (ngu_van) under 2 at Thai Nguyen Examination Council")
print("__________________________________________________________________________________________________________________")

print(df_TN[(df_TN['toan'] < 2) | (df_TN['ngu_van'] < 2)])
print("__________________________________________________________________________________________________________________")

print("i. Draw the histogram of the literature scores of Thai Nguyen Examination Council")
print("__________________________________________________________________________________________________________________")

#task i
plt.hist(df_TN['ngu_van'], bins=10, edgecolor='black')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Histogram of Math in Thai Nguyen Council')
```
